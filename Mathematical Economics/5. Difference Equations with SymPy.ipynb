```python
import sympy as sp
```

```python
sp.init_printing(use_latex='mathjax')
```

```python
α,t,x,y,z=sp.symbols('α t x y z')
```

```python
f=sp.Function('f')
g=sp.Function('g')
```

# <font face="gotham" color="navy">Problem 1:
$$\large \text{simplify: } \Large\Delta(e^{x}) \, \implies \,\, ? $$

```python
sp.difference_delta(sp.exp(x),x,1)
```

# <font face="gotham" color="navy">Problem 2:
$$ \large y_{x+2}-3y_{x+1}+2y_x=4^x $$

```python
de2=sp.Eq(f(x+2)-3*f(x+1)+2*f(x),4**x)

sp.rsolve(de2,f(x))
```

# <font face="gotham" color="navy">Problem 3:
$$ \large y_{t+1}-2y_{t}+5y_{t-1}=30 $$

```python
de3=sp.Eq(f(t+1)-2*f(t)+5*f(t-1),30)

sp.rsolve(de3,f(t))
```

# <font face="gotham" color="navy">Problem 4:
$$ \large y_{x+3}+y_{x+2}-y_{x+1}-y_{x}=0 $$

```python
de4=sp.Eq(f(x+3)+f(x+2)-f(x+1)-f(x),0)

sp.rsolve(de4,f(x))
```

# <font face="gotham" color="navy">Problem 5:
$$ \large y_{t}-2cosh(\alpha)y_{t-1}+y_{t-2}=0 $$

```python
de5=sp.Eq(f(t)-2*sp.cosh(α)*f(t-1)+f(t-2),0)

sp.rsolve(de5,f(t))
```

# <font face="gotham" color="navy">Problem 6:
$$\large y_{x+1}-2y_x=5$$ <br> $$ \large i.c.s. \quad y_0=1 $$

```python
de6=sp.Eq(f(x+1)-2*f(x),5)

sp.rsolve(de6,f(x),{f(0):1})
```

# <font face="gotham" color="navy">Problem 7:
$$ \large \begin{cases} S_t=5 Y_t \\ I_t=15\big(Y_t-Y_{t-1}\big) \\ I_t=S_t\end{cases} $$ <br> $$ \large i.c.s. \quad Y(5)=110 $$

```python
de7=sp.Eq(5*f(t),15*(f(t)-f(t-1)))

sp.rsolve(de7, f(t), {f(5):110})
```

# <font face="gotham" color="navy">Problem 8:
$$ \large Y_t=1.1Y_{t-1}-0.65Y_{t-2}+0.17Y_{t-3}-0.04Y_{t-4} $$ <br> $$ \large i.c.s. \quad \begin{cases} Y_0=65 \\ Y_1=70 \\ Y_2=74 \\ Y_3=79\end{cases} $$

```python
de8=sp.Eq(f(t),1.1*f(t-1)-0.65*f(t-2)+0.17*f(t-3)-0.04*f(t-4))

sp.rsolve(de8, f(t), {f(0):65,f(1):70,f(2):74,f(3):79}).evalf(1)
```
