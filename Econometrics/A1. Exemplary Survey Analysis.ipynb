# <font face="gotham" color="navy"> Regression Analysis I

```python
indep='Q2A +'+ 'Q2B +'+ 'Q2C +'+ 'Q2D +'+ 'Q2E +' + 'Q8 +'+ 'Q55 +'+ 'Q56 +'+ 'Q57 +'+ 'Q58 +'+ 'Q59 +'+ 'Q60_GPA +'+ 'Q61_YEAR_IN_UNI +'+ 'Q62 +'+ 'Q63 +'+ 'Q64 +'+ 'Q65 +'+ 'Q66 +'+ 'Q67_INCOME'
```

```python
m11=smf.gls('Total_Score_Before ~ '+indep, data0).fit()

m11.summary()
```

```python
m12=smf.gls('Raftare_Mali_Before ~ '+indep, data0).fit()

m12.summary()
```

```python
m13=smf.gls('Negareshe_Mali_Before ~ '+ indep, data0).fit()

m13.summary()
```
