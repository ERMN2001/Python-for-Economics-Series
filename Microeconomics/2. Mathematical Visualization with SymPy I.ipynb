```python
import sympy as sp
```

```python
sp.init_printing(use_latex='mathjax')
```

# <font face="gotham" color="purple"> Equations & Systems

```python
x, y, z=sp.symbols('x y z')
```

```python
e1=sp.Eq(1-x**3,9)

display(e1)
```

```python
display(e1.rhs)
e1.lhs
```

```python
sp.solve(e1)
```

```python
sp.checksol(e1,x,1-sp.sqrt(3)*sp.I)
```

```python
e2=sp.Eq(x+3*y-6*z,7)
e3=sp.Eq(-4*x+y+z,-2)
e4=sp.Eq(-z+x+5*y,3)

display(e2,e3,e4)
```

```python
sp.solve([e2,e3,e4],[x,y,z])
```

# <font face="gotham" color="purple"> Inequalities

```python
sp.solve(x>=3*x-13).simplify()
```

```python
ne1=sp.Ne(1-x**3,9)

display(ne1)
```

```python
sp.solve(ne1)
```
