<font face="gotham" color="DarkCyan"><b><span style="font-size:32px"> Section 1</span></b><br><br>
    
<div style="text-align: justify;"><font face="gotham"><span style="font-size:17px">In a perfectly competitive market, if the market's supply and demand functions are given by $p=Q+\frac{1}{Q+4}$ and $p=-\sqrt{Q}+40$ respectively, do the following for each part: </span></div>

- <font face="gotham"><span style="font-size:17px">Compute the equilibrium point.   
- <font face="gotham"><span style="font-size:17px">Draw the supply and demand curves.
- <font face="gotham"><span style="font-size:17px"><div style="text-align: justify;"><b>Calculate</b> and <b>plot</b> the <u>consumer surplus (CS)</u>, <u>producer surplus (PS)</u>, <u>deadweight loss (DWL, if any)</u>, , and <u>government revenue (cost, if any)</u>.</div> <br>

<font face="gotham"><span style="font-size:17px">**Part a)** With no Government Intervention. <br><br>
<font face="gotham">**Part b)** Government imposes a price ceiling equal to 10% of the equilibrium price. <br><br>
<font face="gotham">**Part c)** Government imposes a price floor equal to 10% of the equilibrium price.<br><br>
<font face="gotham">**Part d)** Government levies a lump-sum tax of 5 units on production.<br><br>
<font face="gotham">**Part e)** Government imposes a quota equal to half of the equilibrium quantity.<br><br>
<font face="gotham">**Part f)** Combine all your plotted graphs (except "Part a") into a 2 by 2 grid using the `PlotGrid` command.

<br>
<br>
<font face="gotham" color="OrangeRed"><b><span style="font-size:32px"> Solution</span></b>
<br>
<br>
<br>

```python
import sympy as sp
import sympy.plotting as spp
```

```python
sp.init_printing(use_latex='mathjax')
```

```python
q, p=sp.symbols('q p')
```

<br>
<font face="gotham" color="OrangeRed"><b><span style="font-size:20px"> Part a) [No Policy] </span></b>
<br>
<br>

```python
sup=q+1/(q+4)
dem=40-sp.sqrt(q)
```

```python
qes1=sp.solve(sp.Eq(dem,sup),q) ; qe1=qes1[2].evalf(4)
pe1=sup.subs(q,qe1)

display("Quantity",qe1,'Price',pe1)
```

```python
Dp1=spp.plot_implicit(sp.Eq(p,dem),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='red', title='With No Government Intervention', show=False)

Sp1=spp.plot_implicit(sp.Eq(p,sup),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='blue', show=False)

L11=spp.plot_implicit(sp.Eq(q,qe1),(q,0,50),(p,0,pe1),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L12=spp.plot_implicit(sp.Eq(p,pe1),(q,0,qe1),(p,0,50),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

CSp1=spp.plot_implicit(sp.And(dem>=p,p>=pe1,q>=0),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='Gold', show=False)

PSp1=spp.plot_implicit(sp.And(p>=sup,p<=pe1,q>=0),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='Darkseagreen', show=False)


Dp1.extend(Sp1)
Dp1.extend(L11); Dp1.extend(L12)
Dp1.extend(CSp1); Dp1.extend(PSp1)

Dp1.show()
```

```python
CS1=sp.integrate(dem,(q,0,qe1))-qe1*pe1
PS1=qe1*pe1-sp.integrate(sup,(q,0,qe1))

display("Consumer Surplus",CS1,"Producer Surplus",PS1.evalf(5))
```

<br>
<font face="gotham" color="OrangeRed"><b><span style="font-size:20px"> Part b) [Price Ceiling] </span></b>
<br>
<br>

```python
pe2p=(pe1*0.9).evalf(4)

qes2=sp.solve(sp.Eq(pe2p,sup))
qe2=qes2[1].evalf(4)

pe2c=dem.subs(q,qe2)

display('Quantity',qe2,'Consumption Price',pe2c,'Production Price',pe2p)
```

```python
Dp2=spp.plot_implicit(sp.Eq(p,dem),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='red', title="After a 10% Price Ceiling Policy", show=False)

Sp2=spp.plot_implicit(sp.Eq(p,sup),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='blue', show=False)

L21=spp.plot_implicit(sp.Eq(q,qe2),(q,0,50),(p,0,pe2c),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L22=spp.plot_implicit(sp.Eq(p,pe2p),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='black', show=False)

L23=spp.plot_implicit(sp.Eq(p,pe2c),(q,0,qe2),(p,0,50),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L24=spp.plot_implicit(sp.Eq(q,qe1),(q,0,50),(p,0,pe1),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

CSp2=spp.plot_implicit(sp.And(dem>=p,p>=pe2p,q<=qe2,q>=0),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='Gold', show=False)

PSp2=spp.plot_implicit(sp.And(p>=sup,p<=pe2p,q>=0),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='Darkseagreen', show=False)

DWLp2=spp.plot_implicit(sp.And(dem>=p,p>=sup,q>=qe2,q<=qe1),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='black', show=False)


Dp2.extend(Sp2)
Dp2.extend(L21); Dp2.extend(L22); Dp2.extend(L23); Dp2.extend(L24)
Dp2.extend(CSp2); Dp2.extend(PSp2); Dp2.extend(DWLp2)

Dp2.show()
```

```python
CS2=sp.integrate(dem,(q,0,qe2))-qe2*pe2p
PS2=qe2*pe2p-sp.integrate(sup,(q,0,qe2))
#DWL2=sp.integrate(dem,(q,qe2,qe1))-sp.integrate(sup,(q,qe2,qe1))

display("Consumer Surplus",CS2,"Producer Surplus",PS2.evalf(5),"Dead Weight Loss",(CS1+PS1)-(CS2+PS2))
```

```python
Dp3=spp.plot_implicit(sp.Eq(p,dem),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='red', title="After a 10% Price Floor Policy", show=False)

Sp3=spp.plot_implicit(sp.Eq(p,sup),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='blue', show=False)

L31=spp.plot_implicit(sp.Eq(q,qe3),(q,0,50),(p,0,pe3c),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L32=spp.plot_implicit(sp.Eq(p,pe3p),(q,0,qe3),(p,0,50),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L33=spp.plot_implicit(sp.Eq(p,pe3c),(q,0,50),(p,0,50),
                     axis_center=(0,0), adaptive=False,
                     line_color='black', show=False)

L34=spp.plot_implicit(sp.Eq(q,qe1),(q,0,50),(p,0,pe1),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

CSp3=spp.plot_implicit(sp.And(dem>=p,p>=pe3c,q<=qe3,q>=0),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='gold', show=False)

PSp3=spp.plot_implicit(sp.And(p>=sup,p<=pe3c,q<=qe3,q>=0),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='Darkseagreen', show=False)

DWLp3=spp.plot_implicit(sp.And(dem>=p,p>=sup,q>=qe3,q<=qe1),(q,0,50),(p,0,50),
                     axis_center=(0,0),
                     line_color='black', show=False)


Dp3.extend(Sp3)
Dp3.extend(L31); Dp3.extend(L32); Dp3.extend(L33); Dp3.extend(L34)
Dp3.extend(CSp3); Dp3.extend(PSp3); Dp3.extend(DWLp3)

Dp3.show()
```

<font face="gotham" color="DarkCyan"><b><span style="font-size:32px"> Section 2</span></b><br><br>

<div style="text-align: justify;"><font face="gotham"><span style="font-size:17px">Consider an Internet service provider with a total cost function given by $TC=4q^{3}+10q+5$ in a monopolistic market where demand for the monopolist's services is $p=100-3q^{2}$.</span><font face="gotham"><span style="font-size:17px"> The government has declared that all internet service providers must block price their services for their respective consumers. The monopolist decides to split the remaining demand equally into two price tiers to expand its market for both "heavy internet users" and "light internet users". In this case:</div>

- <font face="gotham"><span style="font-size:17px"> <b>Draw</b> the relevant graphs.</span>
- <font face="gotham"><span style="font-size:17px"> <b>Calculate</b> and <b>plot</b> <u>consumer surplus</u>, <u>producer surplus</u>, <u>lost welfare (if any)</u>, and <u>government revenue (cost, if any)</u>.</span>

```python
TC=4*q**3+10*q+5
D=100-3*q**2

MC=TC.diff(q)
AC=TC/q
MR=(D*q).diff(q)

display('Total Cost',TC,'Demand',D,'Marginal Cost',MC,'Average Cost',AC,'Marginal Revenue',MR)
```

```python
qems=sp.solve(sp.Eq(MR,MC),q); qem=qems[1].evalf(3)
pem=D.subs(q,qem)

qecs=sp.solve(sp.Eq(D,MC),q); qec=qecs[1].evalf(3)
pec=D.subs(q,qec)

display('Monopolistic Quantity',qem,'Monopolistic Price',pem,'Competitive Quantity',qec,'Competitive Price',pec)
```

```python
Dp=spp.plot_implicit(sp.Eq(p,D),(q,0,6),(p,0,110),
                     axis_center=(0,0), adaptive=False,
                     line_color='red', show=False)

MCp=spp.plot_implicit(sp.Eq(p,MC),(q,0,6),(p,0,110),
                     axis_center=(0,0), adaptive=False,
                     line_color='blue', show=False)

ACp=spp.plot_implicit(sp.Eq(p,AC),(q,0.0001,6),(p,0,110),
                     axis_center=(0,0), adaptive=False,
                     line_color='Royalblue', show=False)

MRp=spp.plot_implicit(sp.Eq(p,MR),(q,0,6),(p,0,110),
                     axis_center=(0,0), adaptive=False,
                     line_color='Crimson', show=False)

L1=spp.plot_implicit(sp.Eq(q,qem),(q,0,6),(p,0,pem),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L2=spp.plot_implicit(sp.Eq(p,pem),(q,0,qem),(p,0,110),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L3=spp.plot_implicit(sp.Eq(q,qec),(q,0,6),(p,0,pec),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L4=spp.plot_implicit(sp.Eq(p,pec),(q,0,qec),(p,0,110),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L5=spp.plot_implicit(sp.Eq(q,qem+(qec-qem)/2),(q,0,6),(p,0,D.subs(q,qem+(qec-qem)/2)),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

L6=spp.plot_implicit(sp.Eq(p,pec+(pem-pec)/2),(q,0,qem+(qec-qem)/2),(p,0,110),
                     axis_center=(0,0), adaptive=True,
                     line_color='lightsteelblue', show=False)

FisrtBlockp=spp.plot_implicit(sp.Eq(p,pec+(pem-pec)/2),(q,qem,qem+(qec-qem)/2),(p,0,110),
                     axis_center=(0,0), adaptive=True,
                     line_color='DarkViolet', show=False)

SecondBlockp=spp.plot_implicit(sp.Eq(p,pec),(q,qem+(qec-qem)/2,qec),(p,0,110),
                     axis_center=(0,0), adaptive=True,
                     line_color='DarkViolet', show=False)

CSp1=spp.plot_implicit(sp.And(D>=p,p>=pem,q<=qem,q>=0),(q,0,6),(p,0,110),
                     axis_center=(0,0),
                     line_color='gold', show=False)

PSp1=spp.plot_implicit(sp.And(MC<=p,p<=pem,q<=qem,q>=0),(q,0,6),(p,0,110),
                     axis_center=(0,0),
                     line_color='Darkseagreen', show=False)

CSp2=spp.plot_implicit(sp.And(D>=p,p>=D.subs(q,qem+(qec-qem)/2),q>=qem,q<=qem+(qec-qem)/2,q>=0),(q,0,6),(p,0,110),
                     axis_center=(0,0),
                     line_color='goldenrod', show=False)

PSp2=spp.plot_implicit(sp.And(MC<=p,p<=D.subs(q,qem+(qec-qem)/2),q>=qem,q<=qem+(qec-qem)/2,q>=0),(q,0,6),(p,0,110),
                     axis_center=(0,0),
                     line_color='limegreen', show=False)

CSp3=spp.plot_implicit(sp.And(D>=p,p>=pec,p<=D.subs(q,qem+(qec-qem)/2),q>=qem+(qec-qem)/2,q<=qec,q>=0),(q,0,6),(p,0,110),
                     axis_center=(0,0),
                     line_color='orange', show=False)

PSp3=spp.plot_implicit(sp.And(MC<=p,p<=pec,q>=qem+(qec-qem)/2,q<=qec,q>=0),(q,0,6),(p,0,110),
                     axis_center=(0,0),
                     line_color='lime', show=False)


Dp.extend(MCp); Dp.extend(ACp); Dp.extend(MRp) 
Dp.extend(L1); Dp.extend(L2); Dp.extend(L3); Dp.extend(L4); Dp.extend(L5); Dp.extend(L6)
Dp.extend(FisrtBlockp); Dp.extend(SecondBlockp)

Dp.extend(CSp1); Dp.extend(CSp2); Dp.extend(CSp3)
Dp.extend(PSp1); Dp.extend(PSp2); Dp.extend(PSp3)

Dp.show()
```

```python
CS01=(sp.integrate(D,(q,0,qem))-qem*pem)
CS02=(sp.integrate(D,(q,qem,qem+(qec-qem)/2))-((qec-qem)/2)*(D.subs(q,qem+(qec-qem)/2)))
CS03=(sp.integrate(D,(q,qem+(qec-qem)/2,qec))-((qec-qem)/2)*(pec))

PS01=(qem*pem-sp.integrate(MC,(q,0,qem)))
PS02=(((qec-qem)/2)*(D.subs(q,qem+(qec-qem)/2)))-(sp.integrate(MC,(q,qem,qem+(qec-qem)/2)))
PS03=(((qec-qem)/2)*pec-(sp.integrate(MC,(q,qem+(qec-qem)/2,qec))))

CS=CS01+CS02+CS03
PS=PS01+PS02+PS03

display("Consumer Surplus",CS,"Producer Surplus",PS)
```
