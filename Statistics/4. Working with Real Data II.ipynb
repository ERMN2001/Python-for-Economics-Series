# <font face="gotham" color="navy"> Getting Started

```python
import pandas as pd
import seaborn as sns     # for heatmap
import matplotlib.pyplot as plt
```

```python
plt.rcParams["mathtext.fontset"] = "stix"
plt.rcParams["font.family"] = "STIXGeneral"
```

# <font face="gotham" color="navy"> Data Cleaning

```python
gdp0 = pd.read_excel('P_Data_Extract_From_World_Development_Indicators.xlsx', index_col='Series Name')

gdp1=gdp0.dropna(inplace=False)
gdp2=gdp1.drop(["Country Name",'Country Code', 'Series Code'],axis=1, inplace=False)
gdp3=gdp2.T
```

# <font face="gotham" color="navy"> Merging Plots

```python
plt.figure(figsize=(16,7))
plt.plot(range(2005,2020), gdp3["Tax revenue (% of GDP)"]
         , color='red', label='Tax', lw=1.5)
plt.grid(ls = ':', alpha = 0.25)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Tax', fontsize=24)
plt.xlabel('Year', fontsize=20)

plt.legend(fontsize=18,frameon=True)

plt.show()
```

```python
plt.figure(figsize=(16,7))
plt.plot(range(2005,2020), gdp3["Trade (% of GDP)"]
         , color='blue', label='Trade', lw=1.5)
plt.grid(ls = ':', alpha = 0.25)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Trade vs. Tax', fontsize=24)
plt.xlabel('Year', fontsize=20)

plt.legend(fontsize=18,frameon=True)



plt.plot(range(2005,2020), gdp3["Tax revenue (% of GDP)"]
         , color='red', label='Tax', lw=1.5)

plt.legend(fontsize=18,frameon=True)

plt.show()
```

# <font face="gotham" color="navy"> Scatter Plot

```python
plt.figure(figsize=(10,5))

plt.scatter(gdp3["Tax revenue (% of GDP)"]
            , gdp3["Rural population"]
            , color='red', marker='x')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Scatter Plot', fontsize=20)

plt.show()
```

# <font face="gotham" color="navy"> Correlation Matrix

```python
gdp3.corr()
```

# <font face="gotham" color="navy"> Heatmap

```python
corr = gdp3.corr()

plt.figure(figsize=(18,18))

sns.heatmap(corr, annot=True, cmap='coolwarm', annot_kws={'fontsize': 20})

plt.xticks(fontsize=11.5)
plt.yticks(fontsize=11.5)

plt.xlabel('')
plt.ylabel('')
plt.title('Correlation Heatmap', fontsize=24)

plt.show()
```

# <font face="gotham" color="navy"> Subplots

```python
plt.figure(figsize=(15,8))

plt.subplot(2,3,2)
plt.plot(range(2005,2020), gdp3["Tax revenue (% of GDP)"], color='red', label='Tax', lw=1.5)

plt.subplot(2,3,3)
plt.plot(range(2005,2020), gdp3["Trade (% of GDP)"], color='blue', label='Trade', lw=1.5)
plt.plot(range(2005,2020), gdp3["Tax revenue (% of GDP)"], color='red', label='Tax', lw=1.5)


plt.subplot(2,3,4)
plt.scatter(gdp3["Tax revenue (% of GDP)"]
            , gdp3["Rural population"]
            , color='red', marker='x')

plt.subplot(2,3,6)
sns.heatmap(corr, annot=True, cmap='plasma')

plt.show()
```
