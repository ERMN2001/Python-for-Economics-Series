# <center> Importing Libraries

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# <center> Plotting Discrete Distributions

```python
dist1 = np.random.binomial(10,0.65,100) 

plt.hist(dist1, bins=range(15), density=True, color='blue')

plt.xlabel('Values')
plt.ylabel('Probability')

plt.show()

#plt.savefig('Binomial.svg', dpi=800)
```

```python
dist2 = np.random.poisson(5,1000) 

plt.hist(dist2, bins=range(15), density=True, color='green')

plt.xlabel('Values')
plt.ylabel('Probability')

plt.show()
```

```python
dist3 = np.random.geometric(0.65,100)  

plt.hist(dist3, bins=range(10), density=True, color='red')

plt.xlabel('Values')
plt.ylabel('Probability')

plt.show()
```

```python
dist4 = np.random.hypergeometric(5,10,8,100)

plt.hist(dist4, bins=range(10), density=True, color='coral')

plt.xlabel('Values')
plt.ylabel('Probability')

plt.show()
```

# <center> Plotting Continuous Distributions

```python
x1 = np.random.uniform(-10, 10, 100)

y1 = stats.uniform.pdf(x1, loc=-90, scale=1)

plt.plot(x1, y1,color='chocolate',lw=3)

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()

plt.hist(x1, bins=range(-10,11), density=True, color='chocolate')

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```

```python
x2 = np.random.normal(0,10,1000) 

plt.hist(x2, bins=range(-50,51), density=True, color='palegreen')

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```

```python
x3 = np.random.exponential(6,10000)

plt.hist(x3, bins=range(0,30), density=True, color='slateblue')

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```

```python
x4 = np.random.gamma(2,3,100)

plt.hist(x4, bins=range(0,40), density=True, color='slateblue')

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```

```python
x5 = np.random.beta(2,3,1000)

plt.hist(x5, bins=600, density=True, color='teal')

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```

```python
x6 = np.random.chisquare(21,100)

plt.hist(x6, bins=50, density=True, color='wheat')

plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```
